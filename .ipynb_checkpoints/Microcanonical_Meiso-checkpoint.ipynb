```python
from __future__ import division
import numpy as np
from ambhas.errlib import rmse
from scipy.optimize import fmin
from scipy.stats import poisson
import pandas as pd

def aggrigation(rf):
       
        len_rf = len(rf)
                  
        # take only that much length of rainfall series which is multiplication of 32 
        rf_1 = rf[0:-np.mod(len_rf,32)]
        
        # summing rainfall for 2 days
        rf_2 = np.sum(rf_1.reshape(-1,2), axis=1)
        
        # summing rainfall for 4 days
        rf_4 = np.sum(rf_2.reshape(-1,2), axis=1)
        
        # summing rainfall for 8 days
        rf_8 = np.sum(rf_4.reshape(-1,2), axis=1)
        
        # summing rainfall for 16 days
        rf_16 = np.sum(rf_8.reshape(-1,2), axis=1)
        
        # summing rainfall for 32 days
        rf_32 = np.sum(rf_16.reshape(-1,2), axis=1)
        
        return rf_1,rf_2,rf_4, rf_8,rf_16,rf_32
        
        
        
    
# obtain the data from excel
        
data=pd.read_excel('45_min_Meiso.xlsx')

min_45=data['45 aggrigated Precipitaton']         # The data is in data frame


min_45=np.array(min_45)        # Changing to array

rf_1 =aggrigation(min_45)[0]   # Scale 1/2^5

rf_2 =aggrigation(min_45)[1]   # Scale 1/2^4

rf_4 =aggrigation(min_45)[2]   # Scale 1/2^3

rf_8 =aggrigation(min_45)[3]   # Scale 1/2^2

rf_16 =aggrigation(min_45)[4]  # Scale 1/2^1

rf_32 =aggrigation(min_45)[5]  # Scale 1/2^0
# print(rf_32)
np.mod(len(rf_32),32)
```

# P_01

```python
# import important library

import numpy as np

# example of data

data=rf_1

# breakdown coefcients
def P_01(data):
    k=np.zeros(len(data))

    for i in range(0,len(data),2):
        if i and i+1 ==0:
        
        # for consiquative zeros intenisty 
           
            k[i]=np.nan
        else:
            k[i]=(data[i])/(data[i] + data[i+1])

    # Remove the even zeros  

    even_i = [] 
    k_modefied = [] 

    for i in range(0, len(k)): 
        if i % 2: 
            even_i.append(k[i]) 
        else : 
            k_modefied.append(k[i]) 
          

    # Change the lists to array and remove the nans

    k_modefied=np.asarray(k_modefied)

    k_modefied = k_modefied[np.logical_not(np.isnan(k_modefied))]

    # P_01 is The probability of the breakdown coefficient to be zero or 1.
    
    P_01 = (np.count_nonzero(k_modefied == 1,0))/len(k_modefied)
    
    return P_01
```

```python
P_01_45min_90min=P_01(rf_1)   # P01 45min to 90min which is used to disaggregate from 90 min to 45 min
P_01_1_5hr_3hr=P_01(rf_2)     # P01 1.5hr to 3hr
P_01_3hr_6hr=P_01(rf_4)       # P01 3hr   to 6hr
P_01_6hr_12hr=P_01(rf_8)      # P01 6hr   to 12hr
P_01_12hr_24hr=P_01(rf_16)    # P01 12hr  to 24hr

print (P_01_45min_90min,P_01_1_5hr_3hr,P_01_3hr_6hr,P_01_6hr_12hr,P_01_12hr_24hr)
```

# Probability from 0 to 1   "0<p<1"

```python
def for_P_0_to_1(data):
    k=np.zeros(len(data))

    for i in range(0,len(data),2):
        if i and i+1 ==0:
        # for consiquative zeros intenisty 
            k[i]=np.nan
        else:
            k[i]=(data[i])/(data[i] + data[i+1])

    # Remove the even zeros  

    even_i = [] 
    k_modefied = [] 

    for i in range(0, len(k)): 
        if i % 2: 
            even_i.append(k[i]) 
        else : 
            k_modefied.append(k[i]) 
          

    # Change the lists to array and remove the nans

    k_modefied=np.asarray(k_modefied)

    k_modefied = k_modefied[np.logical_not(np.isnan(k_modefied))]
    
    # P01 is the probability of the breakdown coefficient being zero or one.
    
    k_modefied = k_modefied[k_modefied != 0]
    
    k_modefied = k_modefied[k_modefied != 1]

    return k_modefied 
```

# Disaggregating the data


```python
# generate random variable r.v1

import numpy as np

def microcanonical_beta_dis(R,beta,P_01):
    
    R_dis=np.zeros(2*len(R))
    
    for i in range(0,len(R)*2,2):
        
        # Generate the first random number
        
        r1 = np.random.uniform(size=1)
        if r1<=P_01:
            
            # Generate the second random number
            
            r2=np.random.uniform(size=1)
            
            if r2<= 0.5:
                w1=1
                w2=0
            else:
                w1=0
                w2=1
            R_dis[i]=w1*R[i//2]
            R_dis[i+1]=w2*R[i//2]
        
        else:
            w1=oat(np.random.beta(beta, beta))
            w2=1-w1
            R_dis[i]=w1/(w1+w2)*R[i//2]
            R_dis[i+1]=w2/(w1+w2)*R[i//2]
    
    return R_dis
        
    
```
